```python
def process_data(data, source_lang, destination_lang, output_file):
    count=0
    with open(output_file, 'a') as f:
        writer = csv.writer(f)
        for i in tqdm(range(len(data)),desc ="Translating..."):
            txt=data.iloc[i, 2]
            label=data.iloc[i, 0]
            t_id= data.iloc[i, 1]
            lang=destination_lang
            count=count+1
            if txt:
                try:
                    try:
                        translation= translator.translate(txt, dest=destination_lang, src=source_lang)
                        translated_text=translation.text
                        writer.writerow([label,t_id,translated_text, destination_lang])
                    except Exception as e:
                        print(f"Error Translating sentence 1st method'{count}' '{txt}': {e}\n")
                        print("will try next translator...\n")
                        try:
                            translated_text= GoogleTranslator(source=source_lang, target=destination_lang).translate(text=txt)
                            writer.writerow([label,t_id,translated_text, destination_lang])
                        except Exception as e:
                            print(f"Error Translating sentence 2nd method '{count}' '{txt}': {e}\n")
#                             raise
                except Exception as e:
                    writer.writerow([label,t_id,txt, source_lang])
                    print("Both methods failed, will try again after 10 mint...\n")
                    time.sleep(600)
```

```python
def parallel_processing(file,destination_lang, source_lang):
    filename = str(file)
    destination_lang = str(destination_lang)
    try:
        source_lang = str(source_lang)
    except:
        source_lang = 'auto'
        print('using default source language as auto')

    file_n=filename.split('.')
    output_file=file_n[0]+'_'+destination_lang+'.csv'
    reader = pd.read_csv(filename,encoding = 'latin', chunksize=100000)
    pool = mp.Pool(mp.cpu_count())

    funclist = []
    for df in reader:
        f = partial(process_data, source_lang=source_lang, destination_lang=destination_lang, output_file=output_file)
        funclist.append(pool.apply_async(f, [df]))

    for f in funclist:
        f.get()

    pool.close()
    pool.join()
```
