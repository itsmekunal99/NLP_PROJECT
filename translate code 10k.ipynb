```python
df = pd.read_csv('/content/part_11.csv', encoding = 'latin-1')
```

```python
df.drop(df.columns[0], axis=1, inplace=True)
```

```python
df.head()
```

```python
sentences = df['content'].tolist()
tweet_id = df['id'].tolist()
```

```python
len(sentences)
```

```python
# RUN ONLY ONCE
translated_sentences2 = []
```

```python
for sentence in tqdm(sentences):
    if sentence:  # This checks that sentence is not None or empty
        try:
            translated = GoogleTranslator(source='auto', target='gu').translate(text=sentence) #target is the language code
            translated_sentences2.append(translated)
        except Exception as e:
            print(f"Error translating sentence '{sentence}': {e}")
```

```python
len(translated_sentences2)
```

```python
new_df_2 = pd.DataFrame({'text': translated_sentences2})
```

```python
new_df_2.to_csv('new_df_10k_gu_11.csv', index=False) #DONT FORGET TO WRITE WHICH LANGUAGE IT IS
```
